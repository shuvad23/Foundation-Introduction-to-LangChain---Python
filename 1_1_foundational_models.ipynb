LangChain is a Python/JavaScript framework for building applications powered by Large Language Models (LLMs) like GPT, Gemini, Claude, etc., where the model can:
--
1. Use external tools

2. Read documents

3. Remember context

4. Make decisions

5. Work in chains or agents

👉 In short: " LangChain helps you build intelligent, reasoning-based AI applications, not just chatbots."

---

***Core Components of LangChain***:
1️⃣ LLMs (Language Models)

2️⃣ Prompts & Prompt Templates

3️⃣ Chains (The Heart of LangChain)

4️⃣ Memory (Conversation Awareness)

5️⃣ Tools (LLM Can Take Actions)

6️⃣ Agents (Autonomous Decision Makers)

7️⃣ Document Loaders

8️⃣ Text Splitters

9️⃣ Embeddings

🔟 Vector Stores (Knowledge Base)

---

***🏗️ LangChain Architecture (High Level)***

User
 ↓
PromptTemplate
 ↓
Chain / Agent
 ↓
LLM
 ↓
Tools / Memory / Vector DB
 ↓
Response

### Streaming Output

```python
for token, metadata in agent.stream(
    {"messages": [HumanMessage(content="Tell me all about Luna City, the capital of the Moon")]},
    stream_mode="messages"
):
    # token is a message chunk with token content
    # metadata contains which node produced the token

    if token.content:  # Check if there's actual content
        print(token.content, end="", flush=True)  # Print token
```

#Basic prompting

```python
from langchain.messages import HumanMessage

agent = create_agent(llm)
question = HumanMessage(content="what is 2026 Davos update")
response = agent.invoke({"messages": [question]})
print(response['messages'][1].content)
```

```python
system_prompt = """You are a professional AI assistant with up-to-date knowledge of the 2026 World Economic Forum Annual Meeting (WEF Davos 2026). Use verified news and updates from January 2026.

When answering queries related to Davos 2026, provide:
- Key themes and major outcomes from the event (geopolitical tensions, AI governance, economic priorities).
- Who attended and participated (world leaders, CEOs, global institutions).
- Notable initiatives, debates, agreements, or controversies that emerged.
- Concise, balanced summaries citing credible sources (e.g., Reuters, AP, WEF official releases).
- Clarify if information is uncertain, in progress, or speculative.

Avoid making unverified claims or projecting long-term outcomes that aren’t supported by reliable reporting as of January 2026.

Answer in clear, concise language suitable for general readers, professionals, and analysts.
"""
economic_agent= create_agent(
    model=llm,system_prompt=system_prompt
)

response = economic_agent.invoke({"messages": [question]})
print(response['messages'][1].content)
```

#Structured output

```python
from langchain.agents import create_agent
from langchain.messages import HumanMessage
from pydantic import BaseModel

class CapitalInfo(BaseModel):
    name: str
    location: str
    vibe: str
    economy: str

agent = create_agent(
    model=llm,
    system_prompt="You are a science fiction writer, create a capital city at the users request.",
    response_format=CapitalInfo
)

question = HumanMessage(content="What is the capital of The Moon?")

response = agent.invoke(
    {"messages": [question]}
)

response["structured_response"]
```

```python
response['structured_response'].name
```

```python
response['structured_response'].location
```

```python
response['structured_response'].economy
```

```python
capital_info = response["structured_response"]

capital_name = capital_info.name
capital_location = capital_info.location

print(f"{capital_name} is a city located at {capital_location}")
```
